```python
import pandas as pd
import numpy as np

file_path = "Sô liệu Ly 2.xlsx"
df = pd.read_excel(file_path, sheet_name="Sheet2")

df.columns = df.columns.str.strip()
df.replace("NULL", np.nan, inplace=True)

# Loại bỏ các cột theo yêu cầu
columns_to_drop = [
    "SoBenhAn",
    "SoTiepNh",
    "TenDichVu",
    "MaICD Ngoại trú",
    "Unnamed: 7",
]
df = df.drop(columns=[col for col in columns_to_drop if col in df.columns])

cond_chinh = df["MaICD chính"].astype(str).str.conta("I50|i50", na=False)
cond_phu = df["MaICD phụ"].astype(str).str.conta("I10|i10", na=False)

df_filtered = df[cond_chinh & cond_phu].copy()

for col in ["thoigianvaovien", "thoigianravien", "NgayThucHien"]:
    if col in df_filtered.columns:
        df_filtered[col] = pd.to_datetime(df_filtered[col], errors="coerce")
        df_filtered[col] = df_filtered[col].dt.strftime("%Y-%m-%d %H:%M:%S")

df_filtered["MaICD chính"] = df_filtered["MaICD chính"].str.upper()
df_filtered["MaICD phụ"] = df_filtered["MaICD phụ"].str.upper()

df_filtered.fillna("KHÔNG XÁC ĐỊNH", inplace=True)

print(f"Số lượng dòng giữ lại sau khi lọc: {len(df_filtered)}")
display(df_filtered.head())

df_filtered.to_csv("Cleaned_Data.csv", index=False, encoding="utf-8-sig")
```
